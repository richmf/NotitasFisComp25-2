```python
import math as m
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
y = sp.airy(x)[0]
plt.plot(x,y)
plt.title("Funcionalistas de Airy")
plt.grid()
plt.show()
```

```python
def mi_airy(x):
    return sp.airy(x)[0]
```

```python
def buscar_raiz(f,x_izq,x_der,err=1e-8):
    f_izq , f_med = f(x_izq) , 2*err
    d_raiz = 2*err
    #contador = 0
    while abs(f_med) >= err or d_raiz >= err:
        x_med = 0.5*(x_izq+x_der)
        f_med = f(x_med)
        fifm = f_izq*f_med
        if fifm < 0:
            x_izq , x_der = x_izq , x_med
        else:
            x_izq , x_der = x_med , x_der
            f_izq = f_med
        #contador +=1
        d_raiz = abs(x_izq-x_der)
        #print(contador, '\t f = ',f_med,'\t x0 = ',x_med)
    return x_med
```

```python
x0 = buscar_raiz(mi_airy,-3,-1)
x0
```

```python
x = np.linspace(-5,5,100)
y = sp.airy(x)[0]
plt.title("Funcionalistas de Airy")
plt.plot(x,y,label=r"$Ai(x)$")
plt.plot(x0,[0],'x',label=r"ra√≠z $x_0$")
plt.grid(),plt.legend()
plt.show()
```

```python
def mi_pol(x):
    return x**7
```

```python
x0 = buscar_raiz(mi_pol,-2,5)
```

```python
x = np.linspace(-0.5,0.5)
plt.plot(x,mi_pol(x))
plt.plot(x0,[0],'+')
plt.grid()
plt.show()
```
