```python
import math as m
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Vcuadcuad(x,y,L=0.5,V0=1.0):
    Nx , Ny = len(x) , len(y)
    malla = np.zeros((Nx,Ny))
    for i in range(1,Nx-1):
        for j in range(1,Ny-1):
            if  -0.5*L < x[i] < 0.5*L and -0.5*L < y[j] < 0.5*L:
                malla[i,j] = 1
            else:
                malla[i,j] = rd.random()
    return V0*malla

def Relajar(V,V0=1):
    Nx , Ny = V.shape
    for i in range(1,Nx-1):
        for j in range(1,Ny-1):
            if V[i,j] != 1:
                V[i,j] = 0.25*( V[i+1,j] + V[i-1,j] + V[i,j+1] + V[i,j-1] )
    return V

def Met_Jocobi(V,V0=1,err=1e-4):
    diferencia = 1
    while diferencia > err:
        for i in range(20):
            V = Relajar(V,V0=V0)
        V_ref = Relajar(np.copy(V))
        diferencia = np.max(abs(V_ref-V))
    return V
```

```python
x , y = np.linspace(-3,4,100) , np.linspace(-2,2,100)
pot = Vcuadcuad(x,y)
plt.imshow(pot.transpose(),
           extent=(x[0],x[-1],y[0],y[-1]),
           origin='lower',
           cmap='bone')
plt.show()
```

```python
pot = Met_Jocobi(pot)
```

```python
plt.imshow(pot.transpose(),
           interpolation='bilinear',
           extent=(x[0],x[-1],y[0],y[-1]),
           origin='lower',
           cmap='bone')
plt.show()
```

```python
niveles = np.linspace(0,1,10)
plt.contour(x,y,pot.transpose(),
            levels=niveles,
            origin='lower',
            cmap='bone')
plt.gca().set_aspect('equal')
plt.show()
```

```python
Ex,Ey = np.gradient(-pot.transpose())
```

```python
plt.quiver(x,y,Ey,Ex)
plt.contour(x,y,pot.transpose(),
            levels=niveles,
            origin='lower',
            cmap='bone')

plt.show()
```

```python
plt.streamplot(x,y,Ey,Ex,density=1.5)
plt.contour(x,y,pot.transpose(),
            levels=niveles,
            origin='lower',
            cmap='viridis')

plt.show()
```
