```python
import math as m
import random as rd
import numpy as np
import numpy.linalg as al
import matplotlib.pyplot as plt
```

```python
def MatA(x,r):
    A = np.zeros( (len(x),len(x)) )
    dd = 2*(1+r)
    for i in range(len(x)-1):
        A[i,i] = dd # diagonal
        A[i,i+1] = -r # diagonal superior
        A[i+1,i] = -r # diagonal inferior
    A[-1,-1] = dd
    return A

def MatB(x,r):
    B = np.zeros( (len(x),len(x)) )
    dd = 2*(1-r)
    for i in range(len(x)-1):
        B[i,i] = dd # diagonal
        B[i,i+1] = r # diagonal superior
        B[i+1,i] = r # diagonal inferior
    B[-1,-1] = dd
    return B

def MetCR(x,t,Psi0,D=0.5):
    # Nota: Esto sirve solo para los puntos interiores de x y Psi0
    r = D*(t[1]-t[0])/((x[1]-x[0])**2)
    sol = np.zeros( (len(x),len(t)) )
    sol[:,0] = Psi0
    print('r = ',r)
    if r < 1:
        A , B = MatA(x,r) , MatB(x,r)
        Ainv = al.inv(A)
        MatD = np.dot(Ainv,B)
        for i in range(len(t)-1):
            sol[:,i+1] = np.dot(MatD,sol[:,i])
    else:
        print('No hay convergencia')
    return sol
```

```python
x = np.linspace(0,1)
P0 = -x*(x-1)
plt.plot(x,P0)
plt.show()
```

```python
x_int , P0_int = x[1:-1] , P0[1:-1]
t = np.linspace(0,1,1300)
Psi_xt = MetCR(x_int,t,P0_int)
Psi_xt.shape
```

```python
plt.plot(x_int,Psi_xt[:,0])
plt.plot(x_int,Psi_xt[:,100])
plt.plot(x_int,Psi_xt[:,500])
plt.plot(x_int,Psi_xt[:,-1])
plt.show()
```
